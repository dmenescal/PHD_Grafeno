```python
from pathlib import Path
```

# Obtendo os nomes dos arquivos de imagem

```python
data_path = Path('..')
img_files = []

for e in data_path.rglob('*.png'):
    img_files.append(str(e))
```

```python
len(img_files)
```

# Quebra do string para obter os seguintes valores: Im(Kx) e Alpha

```python
imn2_lst = []
imkx_lst = []
loss_lst = []

for img in img_files:
    splited = img.split('_')

    imkx = splited[6]
    loss = splited[8]
    imn2 = splited[4]

    # Im(n2): Remover a string i
    imn2_len = len(imn2)
    imn2 = float(imn2[0:imn2_len-1])

    # Im(Kx): Remover a string i
    imkx_len = len(imkx)
    imkx = float(imkx[0:imkx_len-1])

    # Loss: Remover o .png
    loss_len = len(loss)
    loss = float(loss[0: loss_len-4])
    
    imkx_lst.append(imkx)
    loss_lst.append(loss)
    imn2_lst.append(imn2)
```

```python
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.DataFrame(list(zip(imn2_lst, imkx_lst, loss_lst)), columns=['Im(n2)', 'Im(kx)', 'Loss'])
```

```python
sns.lineplot(data=dataset, x='Im(n2)', y='Im(kx)')
```

```python
sns.lineplot(data=dataset, x='Im(n2)', y='Loss')
```
